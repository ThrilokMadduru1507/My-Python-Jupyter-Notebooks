```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
# Solution for AS Problem Statement Question 1

DATA = {'Planned to purchase Product A':['Yes','No','Total'],'Actually placed order for Product A - Yes':[400,200,600],
       'Actually placed order for Product A - No':[10,1300,1400],'Total':[500,1500,2000]}

DF = pd.DataFrame(DATA)

print(DF)

# Joint probability of the people who planned to purchase and actually placed an order. - SOL1

SOL1 = 400/2000

print("\n\n1.A. Joint probability of the people who planned to purchase and actually placed an order:",SOL1)

# Joint probability of the people who planned to purchase and actually placed an order, 
#given that people planned to purchase.

SOL2 = 400/500

print("\n\n1.B. Joint probability of the people who planned to purchase and actually placed an order,\ngiven that people planned to purchase:",SOL2)
```

```python
# Solution for AS Problem Statement Question 2

#An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. 
#Historically, the failure rate for the manufactured item is 5%. 
#Suppose a random sample of 10 manufactured items is selected.

#FAILURE RATE - FR

FR = 0.05

#RANDOM SAMPLE - RS

RS = 10

#SAMPLE DATA - SD

SD = np.arange(0,10)

#BINOMIAL DISTRIBUTION - BD

BD = stats.binom.pmf(SD,RS,FR)

#Probability that none of the items are defective?

print('\n\n2.A. Probability that none of the items are defective:',BD[0])

#Probability that exactly one of the items is defective?

print('\n\n2.B. Probability that exactly one of the items is defective:',BD[1])

#Probability that two or fewer of the items are defective?

print('\n\n2.C. Probability that two or fewer of the items are defective:',BD[0]+BD[1]+BD[2])

#Probability that three or more of the items are defective ?

print('\n\n2.D. Probability that three or more of the items are defective:',1-(BD[0]+BD[1]+BD[2]+BD[3]))
```

```python
# Solution for AS Problem Statement Question 3

# A car salesman sells on an average 3 cars per week.

R = 3

N = np.arange(0,10)

# POISSON DISTRIBUTION - PD

PD = stats.poisson.pmf(N,R)

#What is Probability that in a given week he will sell some cars?

print('\n\n3.A. What is Probability that in a given week he will sell some cars:',1-PD[0])

#What is Probability that in a given week he will sell 2 or more but less than 5 cars?

SOL = PD[2]+PD[3]+PD[4]

print('\n\n3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars:',SOL)

#Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week.

print('\n\n3.C. Plot the poisson distribution function for cumulative probability of \ncars sold per-week vs number of cars sold per week:')

plt.plot(N,PD, 'o-')
plt.title('Poisson Distribution')
plt.xlabel('Probability of cars sold per-week')
plt.ylabel('Number of cars sold per week')
plt.show()
```

```python
# Accuracy in understanding orders for a speech based bot at a restaurant is important for the Company X which has designed,
# marketed and launched the product for a contactless delivery due to the COVID-19 pandemic. Recognition accuracy that 
# measures the percentage of orders that are taken correctly is 86.8%. Suppose that you place an order with the bot and 
# two friends of yours independently place orders with the same bot. Answer the following questions.

#PROBABILITY OF ORDERS TAKEN CORRECTLY - POTC

POTC = 0.868

#PROBABILITY OF ORDERS NOT TAKEN CORRECTLY - PONTC

PONTC = 1-POTC

# NO OF ORDERS PLACED - NOP

NOP = 3

# SAMPLE ORDERS DATA

SOD = np.arange(0,4)

BINOMINAL = stats.binom.pmf(SOD,NOP,POTC)

print('\n\nBINOMINAL DISTRIBUTION FOR THE PROBLEM STATEMENT: ',BINOMINAL)

# 4.A. What is the probability that all three orders will be recognised correctly?

print('\n\n4.A. What is the probability that all three orders will be recognised correctly?',BINOMINAL[3])

# 4.B. What is the probability that none of the three orders will be recognised correctly?

print('\n\n4.B. What is the probability that none of the three orders will be recognised correctly?',BINOMINAL[0])

# 4.C. What is the probability that at least two of the three orders will be recognised correctly?

print('\n\n4.C. What is the probability that at least two of the three orders will be recognised correctly?',BINOMINAL[2]+BINOMINAL[3])
```

# • DOMAIN: Sports
• CONTEXT: Company X manages the men's top professional basketball division of the American league system. The dataset contains information
on all the teams that have participated in all the past tournaments. It has data about how many baskets each team scored, conceded, how many
times they came within the first 2 positions, how many tournaments they have qualified, their best position in the past, etc.
• DATA DESCRIPTION: Basketball.csv - The data set contains information on all the teams so far participated in all the past tournaments.
• DATA DICTIONARY:
1. Team: Team’s name
2. Tournament: Number of played tournaments.
3. Score: Team’s score so far.
4. PlayedGames: Games played by the team so far.
5. WonGames: Games won by the team so far.
6. DrawnGames: Games drawn by the team so far.
7. LostGames: Games lost by the team so far.

10. TournamentChampion: How many times the team was a champion of the tournaments so far.
11. Runner-up: How many times the team was a runners-up of the tournaments so far.

13. HighestPositionHeld: Highest position held by the team amongst all the tournaments played.
• PROJECT OBJECTIVE: Company’s management wants to invest on proposals on managing some of the best teams in the league. The analytics
department has been assigned with a task of creating a report on the performance shown by the teams. Some of the older teams are already in
contract with competitors. Hence Company X wants to understand which teams they can approach which will be a deal win for them.

```python
# 1. Read the data set, clean the data and prepare final dataset to be used for analysis.

# Read the data set from Basketball.csv into BBDATA of type Pandas DATAFRAME

BBDATA = pd.read_csv("Basketball.csv");

# Create a copy of the Data Set, so that we can have a copy of data available in hand through out the process

# It helps in any modifications done to the data to have a copy of original data available.

BBDATACOPY = BBDATA.copy(deep=True)

print(BBDATA.head())

print('\n\n',BBDATA.tail())

# Tail of the data shows there are some missing values

# Our BBDATA Dataset consists of 61 Rows, each row giving an outline of an individual team

# Information on the BBDATA regarding the attributes and their datatypes

print('\n\n',BBDATA.info())
```

```python
# Cleaning up data

# Let us check whether NULL Values are present in the Data Set.

ISNULL = BBDATACOPY.isna().sum()


if ISNULL.sum()==0:
    print('\n\nNO NULL VALUES ARE PRESENT IN THE DATA')
else:
    print('\n\nTHERE ARE NULL VALUES PRESENT IN THE DATA')
    
```

```python
# Cleaning up data

# Let us check whether DUPLICATE Values are present in the Data Set.

ISDUPE = BBDATACOPY.duplicated().sum()


if ISDUPE==0:
    print('\n\nNO DUPLICATE VALUES ARE PRESENT IN THE DATA')
else:
    print('\n\nTHERE ARE DUPLICATE VALUES PRESENT IN THE DATA')
    
```

```python
# Cleaning up data

# There is special character "-" present in the DATASET in TournamentChampion and Runner-up columns

# Replace them with 0 to make the data look simple.


BBDATA = BBDATA.replace('-',0)

# For Team 61, we do not have all the data details to consider except that it participated once in the tournament and
# got 9th position. For data analysation, we can drop the row pertaining to Team 61 to proceed for futher EDA process.

# TeamLaunch shows between which year, a team is launched and contains special characters -, ~
# We can consider the launch year as first year specified in the data and exclude the part from - or ~
# When analysing the data this helps in understanding and plotting the data easily


BBDATA['TeamLaunch'] = BBDATACOPY['TeamLaunch'].str.slice(0,4)
```

```python
# From BBDATA.info(), we could see that there are numberic values present but treated as Object Data Type.

# We need to change the Columns with numeric data into numeric type for performing statistical operations and plotting

# Since the data contains both Objects and Numeric data

# Store all the column headers into COL_HEAD variable

COL_HEAD = BBDATA.columns.drop('Team')

BBDATA[COL_HEAD] = BBDATA[COL_HEAD].apply(pd.to_numeric, errors = 'coerce')

print('\n\n',BBDATA.tail())
```

# Through Performance matrix we look into abilities, overall quality of the teams and review them by the management to look for any inprovements or suggestions needed to better the performance of a team

Some teams might be launched many years back but might not played in most tournaments, and other side some teams might have launched recently but participated in as much tournaments as they can.

```python
# Let us create 2 samples for companies with operating and closed state

FIRST_SAMPLE = SE_DATA_NEW.loc[SE_DATA_NEW.OperatingState == 'Operating', 'Funds_in_million']

SECOND_SAMPLE = SE_DATA_NEW.loc[SE_DATA_NEW.OperatingState =='Closed', 'Funds_in_million']
```

# Write the null hypothesis and alternative hypothesis.

NULL HYPOTHESIS (Ho) -> NO SIGNIFICANT DIFFERENCE BETWEEN THE 2 MEANS

ALTERNATE HYPOTHESIS (Ha) -> SIGNIFICANT DIFFERENCE IS PRESENT BETWEEN THE 2 MEANS

-> WE ARE GOING TO DO BOTH T-TEST AND Z-TEST TO COMPARE THE TWO SAMPLES CREATED ABOVE AND SEE WHETHER WE CAN REJECT THE NULL HYPOTHESIS OR FAILED TO REJECT

```python
# USING T-TEST METHOD

from scipy.stats import ttest_ind, ttest_ind_from_stats

T_TEST_STATISTIC, P_VALUE = ttest_ind(FIRST_SAMPLE, SECOND_SAMPLE)

print("TTEST_IND:            TEST_STATISTIC = %g  P_VALUE = %g" % (T_TEST_STATISTIC, P_VALUE))

# DEFINE ALPHA VALUE SIGNIFICANCE AT 0.05 or 5%

ALPHA_VALUE = 0.05

if P_VALUE > ALPHA_VALUE:
    print('\n\nSINCE P_VALUE: %g IS GREATER THAN ALPHA_VALUE: %g, WE FAIL TO REJECT NULL HYPOTHESIS'%((round(P_VALUE,2),ALPHA_VALUE)))
else:
    print('\n\nSINCE P_VALUE: %g IS LESS THAN ALPHA_VALUE: %g, WE REJECT THE NULL HYPOTHESIS'%((round(P_VALUE,2),ALPHA_VALUE)))
    
```

```python
# USING Z-TEST METHOD

from statsmodels.stats.weightstats import ztest

from statsmodels.stats.proportion import proportions_ztest

Z_TEST_STATISTIC, P_VALUE1 = ztest(FIRST_SAMPLE, SECOND_SAMPLE)

print("ZTEST_IND:            Z_TEST_STATISTIC = %g  P_VALUE1 = %g" % (Z_TEST_STATISTIC, P_VALUE1))

# DEFINE ALPHA VALUE SIGNIFICANCE AT 0.05 or 5%

ALPHA_VALUE = 0.05

if P_VALUE > ALPHA_VALUE:
    print('\n\nSINCE P_VALUE: %g IS GREATER THAN ALPHA_VALUE: %g, WE FAIL TO REJECT NULL HYPOTHESIS'%((round(P_VALUE1,2),ALPHA_VALUE)))
else:
    print('\n\nSINCE P_VALUE: %g IS LESS THAN ALPHA_VALUE: %g, WE REJECT THE NULL HYPOTHESIS'%((round(P_VALUE1,2),ALPHA_VALUE)))
    
```

Conclusion from Hypothesis testing:

From the above T-Test and Z-Test conducted on the 2 samples with company states as operating and closed, we cant say that a company which raised more funds has been successful or otherwise.

Another scenario might be whether the customer is looking for a credit facility that helps he/she in setting up a new business and make employment out of it or enhance existing business by adding new machinery or additional raw material or agricultural activities or simply a person looking for a temporary financial assistance on certain requirements which can be repaid eventually. Also there might be customers who are looking for additional facilities offered by a bank like purchasing a best insurance, purchasing shares, investment options and many (in simple - para banking options)
